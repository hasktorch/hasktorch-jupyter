Hasktorch Quickstart

1.   Run first bash script, and wait for 30mins.
2.   Change runtime-type to haskell (Click Runtime->Change runtime type-> Save)
3.   Write haskell codes and run it.

Notes: hvega is not available.

```haskell
%%bash
apt -y --allow-downgrades --allow-remove-essential --allow-change-held-packages install locales software-properties-common apt-transport-https
add-apt-repository -y ppa:hvr/ghc
apt-get update -qq 
apt -y --allow-downgrades --allow-remove-essential --allow-change-held-packages \
         install locales software-properties-common apt-transport-https
add-apt-repository -y ppa:hvr/ghc
apt-get update -qq
apt install -y -qq \
        python3-pip libgmp-dev libmagic-dev libtinfo-dev libzmq3-dev \
	      libcairo2-dev libpango1.0-dev libblas-dev liblapack-dev gcc g++ wget git \
	      ghc-8.10.3 cabal-install-3.2 cpphs curl
git clone https://github.com/gibiansky/IHaskell.git ihaskell
git clone https://github.com/hasktorch/hasktorch.git hasktorch
pushd ihaskell && git checkout e07b198f07d7dc7a12863f8507439ae3834947e4 && popd
pushd hasktorch && git checkout be74a1eaabc2a6b56e21bbb24c870299d8e3bdb6 && popd
curl https://www.stackage.org/lts-17.2/cabal.config > cabal.freeze

pushd /usr/bin/; for i in /opt/ghc/bin/* ; do ln -s $i . ;done;popd
pushd /usr/lib/; for i in /usr/local/lib/python3.6/dist-packages/torch/lib/*.so ; do ln -s $i . ;done;popd
pushd /usr/include/; for i in /usr/local/lib/python3.6/dist-packages/torch/include/* ; do ln -s $i . || true;done;popd
pushd /usr/include/torch; for i in csrc/api/include/torch/* ; do ln -s $i .  || true;done;popd
pushd /usr/lib/; for i in /usr/local/lib/python3.6/dist-packages/torch/lib/* ; do ln -s $i . || true;done;popd
pushd /usr/bin/; for i in /root/.cabal/bin/ihaskell ; do ln -s $i . ;done;popd


cabal update
cabal install alex happy

pushd /usr/bin/; for i in /root/.cabal/bin/* ; do ln -s $i . ;done;popd

cabal v1-install \
          ./ihaskell \
          ./ihaskell/ipython-kernel \
          ./ihaskell/ghc-parser \
	        ihaskell-hvega \
          ./hasktorch/hasktorch \
          ./hasktorch/libtorch-ffi \
          ./hasktorch/libtorch-ffi-helper \
          --ghc-options "-j2 +RTS -A128m -n2m -RTS"
ihaskell install
```
